```python
from mpl_toolkits.mplot3d import axes3d
from scipy.special import sph_harm
```

```python
def sph2cart(r, ϕ, θ):
    """
    r is from 0 to infinity 
    ϕ is from 0 to 2*pi 
    θ is from 0 to pi
    """
    x = r* np.sin(θ)* np.cos(ϕ)
    y = r* np.sin(θ)* np.sin(ϕ)
    z = r* np.cos(θ)
    return x, y, z
```

```python
# ϕ running from 0 to 2*π and θ from 0 to π
π   = np.pi
ϕ   = np.linspace(0, 2*π,21)
θ   = np.linspace(0,   π,21)
# meshgrid to generate points
ϕm, θm = np.meshgrid(ϕ, θ)

# Y = np.cos(tta)
Y  = sph_harm(0, 1, θm, ϕm)
Y2 = np.real(np.conj(Y.T)*Y.T)

x, y, z = sph2cart( Y2 , ϕm, θm)

fig = plt.figure()
ax = fig.add_subplot( 111 , projection='3d')

from matplotlib import cm
# ax.set_aspect('equal')
ax.plot_surface(x, y, z, linewidth = 0.5, edgecolors = 'k')

ax.set_xlim(-0.4,0.4)
ax.set_ylim(-0.4,0.4)
ax.set_zlim(-0.4,0.4)
```
